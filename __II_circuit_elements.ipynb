Ankstesniame skyriuje aptarti kai kurie grandinės elementai yra idealūs grandinės elementai.
Realaus pasaulio grandinės elementai yra gana artimi idealiems matematiniams modeliams, tačiau neišvengiamai bus netobuli (neatitks matematiniu modeliu). Geras inžinierius blaiviai įvertina tikrų komponentų ribotumus ir geba palyginti juos su jų idealiomis abstrakcijomis matematiniuose modeliuose.
<br>

Vienas paprastas pavyzdys tai kad nuo idealių verčių fiziniai įrenginiai, tokie kaip rezistoriai, induktoriai ir kondensatoriai
(R, L, C) turi tam tikrą nuokrypį nuo idealios vertės tai vadinama tolerancijos lygis (kuo tolerancija griežtesnė, kuo mažesnis nuokrypis tuo daugiau pinigų kainuoja toks elementas). Tikrieji komponentai niekada neturi tiksliai nurodytų verčių.
<br>

Tikrieji grandinės elementai nukrypsta nuo idealių lygčių,kraštutiniais įtampos arba srovė atvejais .Matematinė rezistoriaus abstrakcija netampa tikrovėje 
begalinė įtampa arba begalinė srovė. Modelis tam tikru momentu sugenda ir komponentas tiesiog sunaikinamass. Visų idealių komponentų ir šaltinių abstraktūs modeliai realiame pasaulyje yra riboti.
<br>
Kalbant apie rezistorių kaip pavyzdį: kadangi laidai, prijungti prie rezistoriaus galų, sukuria aplinkinį magnetinį lauką, rezistorius neišvengiamai parodys tam tikras indukcines savybes. Be to, rezistoriai yra pagaminti iš laidžių medžiagų ir dažniausiai yra šalia kitų laidininkų. Kartu šie laidininkai veikia kaip kondensatoriaus plokštės, todėl rezistoriai taip pat turi tam tikrų talpinių savybių. Šie parazitiniai poveikiai gali būti svarbūs esant aukštiems dažniams arba kai smarkiai keičiasi įtampa ar srovė. Jei parazitiniai reiškinia grandinėje yra svarbūs, galite modeliuoti komponentą kaip idealių elementų derinį, kaip parodyta čia su rezistoriumi:

Realaus pasaulio komponentų savybės yra jautrios jų aplinkai. Dauguma komponentų rodo tam tikrą temperatūros jautrumo laipsnį. Komponentų parametrai svyruoja aukštyn arba žemyn, priklausomai nuo to, kiek karštas ar šaltas yra komponentas. Jei jūsų grandinė turi veikti plačiame temperatūrų diapazone, norėsite sužinoti naudojamų komponentų atsparūmą temperatūroms ir pokyčių poveikį.

## Rezistorius

Kuriant tikrus rezistorius, tikslas yra sukurti komponentą, kuris būtų kuo artimesnis Ohmo dėsnio veikimui.<br>

<div class='alert alert-success'>
<p><b>Omo desnis:</b> 
\begin{multline*}
    V = IR
\end{multline*}
</p>
</div>

Rezistoriaus varžos vertė priklauso nuo dviejų dalykų, tai yra iš ko jis padarytas ir kokios formos jis yra. Tūrinė medžiaga (iš ko rezistorius pagamintas) turi įtakos elektronų tekėjimo laidumui. Galite galvoti apie tai, kaip dažnai elektronai atsitrenkia į medžiagoje esančius atomus, bandydami tekėti pro juos. Ši medžiagų savybė vadinama varža (laidumas, yra atvirkščias varžai). Pasirinkus tam tikros varžos tūrinę medžiagą, rezistoriaus varža nustatoma pagal jo formą. Ilgesnis rezistorius turi didesnę varžą nei trumpesnis, nes elektronai patiria daugiau susidūrimų, kai jie praeina per medžiagos atomus. Didesnio skerspjūvio ploto rezistorius turi mažesnę varžą nei mažesnio skerspjūvio rezistorius, nes elektronai turi daugiau kelių.
  
 - Rezistorius yra grandinės elementas, fizinis objektas.
 - Varža yra medžiagų savybė.
 - Todėl varžą yra rezistoriaus savybė, kurią lemia ir medžiagos iš kurios yra pagamintas rezistorius varža, ir rezistoriaus forma.
 
Tikras rezistorius sugenda (kaip perdega), jei rezistoriaus išsklaidoma galia yra didesnė nei gali atlaikyti jo medžiagos. Rezistoriai turi galią, kurios negalima viršyti. Jei bandysite išsklaidyti 1 vatą galios 1/8 vato rezistoriuje viskas gali baigtis perdegusiu daiktu.

### Žymėjimas

TH modelio rezistoriai yra spalvoti. Spalvos turi reikšmes šios spalvų juostos ant rezistoriaus rodo jų varžos vertę ir toleranciją. Tarkime rezistoriaus juostos yra oranžinė, ruda ir aukso spalvos. Rezistoriaus spalvų kodų diagramoje pirmosios dvi juostos atitinka vertės skaitmenis ,trečioji juosta yra daugiklis.  Ketvirta (paskutinė) juosta rodo toleranciją. Šio konkretaus rezistoriaus vertė yra 10 kΩ ±5%.
