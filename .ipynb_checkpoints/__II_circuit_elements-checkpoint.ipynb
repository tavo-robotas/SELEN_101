## Idealus elementai

Kad kurti ir analizuoti elektros grandines vienas dalykas kuri turime padaryti yra tai kad reikai turėti su kuo daryti šiuos grandynus. Trumpai aptarsime <code>**labai dažnus 3 grandynės <i>pasyvius</i> elementus: rezistorius, kondensatorius, induktorius**</code>. 

- Rezistorius - simbolis yra zigzagas kuris ir reiškia kad srovei yra daugiau darbo kad praeiti pro šia grandies dalį. Žymėjimo simbolis yra <code>**R**</code> 
- Kondensatorius - simbolis reprezentuoja iš ko jis ir yra dažniausiai pagamintas, tai du laidininkai (arba metaliniai objektai) kurie yra šalia vienas kito. Žymėjimo simbolis yra <code>**C**</code> 
- Induktorius - simbolis tai pat yra repreznetuojantis iš ko yra pagaminta tai yra vielos ritės. Žymėjimo simbolis yra <code>**L**</code> 
 todėl kad I (current) jau yra naudojamas žymėti srovei, kaip negalime srovei naudotoi (C). 
 
Visi šie komponentai turi tarytumei su savim lydinčia (iv) lygtį kuri išreiškia ryšį tarp srovės ir įtampos.

 - Itampą per rezisotorių yra proporcinga srovės per rezistorių sandaugai su varžos konstanta (v = iR);
 - Srovė per kondesatorių yra proporcinga įtampos pokyčiui per laiką;
 - Induktoriaus savybė labai panašį į kondensatoriaus tik skirtumas kad įtampa per induktorių yra proporcinga srovės pokyčiui per laiką.

<div class='alert alert-info'>
\begin{multline*}
v = iR
\end{multline*}
<br>   

\begin{multline*}
i = C\dfrac{dv}{dt}
\end{multline*}
<br>    

\begin{multline*}
v = L\dfrac{di}{dt}
\end{multline*}
</div>

Tai pat aptarsime idealu įtampos ir srovės šaltinį. Pastovios įtampos šaltinis gali būti tarkime reguliuojamos įtampos šaltinis arba baterija (baterija turi atskirą žymėjimo simbolį su poliariškumo konvecija).

Idealus kintamos įtampos šaltinis generuoja įtampą kaip laiko funkciją, nepriklausomą nuo srovės, kaip parodyta šioje įtampos ir laiko diagramoje.

<div class='alert alert-info'>
\begin{multline*}
v = v(t)
\end{multline*}
<br>  
Tai v(t) būti sinusinė banga arba bet kuri kita laike kintanti įtampa, pavyzdžiui, viena įtampos pakopa arba pasikartojanti kvadratinė banga ir t.t 
</div>

Kol kas aptarti kai kurie grandinės elementai yra idealūs grandinės elementai.
Realaus pasaulio grandinės elementai yra gana artimi idealiems matematiniams modeliams, tačiau neišvengiamai bus netobuli (neatitks matematiniu modeliu). Geras inžinierius blaiviai įvertina tikrų komponentų ribotumus ir geba palyginti juos su jų idealiomis abstrakcijomis matematiniuose modeliuose.
<br>

Vienas paprastas pavyzdys tai kad nuo idealių verčių fiziniai įrenginiai, tokie kaip rezistoriai, induktoriai ir kondensatoriai
(R, L, C) turi tam tikrą nuokrypį nuo idealios vertės tai vadinama tolerancijos lygis (kuo tolerancija griežtesnė, kuo mažesnis nuokrypis tuo daugiau pinigų kainuoja toks elementas). Tikrieji komponentai niekada neturi tiksliai nurodytų verčių.
<br>

Tikrieji grandinės elementai nukrypsta nuo idealių lygčių,kraštutiniais įtampos arba srovė atvejais .Matematinė rezistoriaus abstrakcija netampa tikrovėje 
begalinė įtampa arba begalinė srovė. Modelis tam tikru momentu sugenda ir komponentas tiesiog sunaikinamass. Visų idealių komponentų ir šaltinių abstraktūs modeliai realiame pasaulyje yra riboti.
<br>
Kalbant apie rezistorių kaip pavyzdį: kadangi laidai, prijungti prie rezistoriaus galų, sukuria aplinkinį magnetinį lauką, rezistorius neišvengiamai parodys tam tikras indukcines savybes. Be to, rezistoriai yra pagaminti iš laidžių medžiagų ir dažniausiai yra šalia kitų laidininkų. Kartu šie laidininkai veikia kaip kondensatoriaus plokštės, todėl rezistoriai taip pat turi tam tikrų talpinių savybių. Šie parazitiniai poveikiai gali būti svarbūs esant aukštiems dažniams arba kai smarkiai keičiasi įtampa ar srovė. Jei parazitiniai reiškinia grandinėje yra svarbūs, galite modeliuoti komponentą kaip idealių elementų derinį, kaip parodyta čia su rezistoriumi:

Realaus pasaulio komponentų savybės yra jautrios jų aplinkai. Dauguma komponentų rodo tam tikrą temperatūros jautrumo laipsnį. Komponentų parametrai svyruoja aukštyn arba žemyn, priklausomai nuo to, kiek karštas ar šaltas yra komponentas. Jei jūsų grandinė turi veikti plačiame temperatūrų diapazone, norėsite sužinoti naudojamų komponentų atsparūmą temperatūroms ir pokyčių poveikį.

Jums nereikės dažnai jaudintis dėl parazitinio poveikio. Jie čia paminėti, kad žinotumėte, kad jie egzistuoja. Kai imituojate elektronines grandines, jums nereikia komplikuoti reikalų modeliuojant visus galimus parazitinius efektus, nebent turite pagrindo manyti, kad jie bus svarbūs.

Įtampa yra tiesiogiai proporcinga srovei, tekančiai per rezistoriu.
<div class='alert alert-success'>

\end{multline*}
<br>

R yra šio proporcingumo konstanta, ir reiškia pasipriešinimą/atsparumą. 
Atsparumas matuojamas omų vienetus, žymima graikų raide  omega simboliu Ω.

## Kondesatorius

Gaminant tikrus kondensatorius, tikslas yra sukurti komponentą, kuris būtų kuo arčiau idealaus kondensatoriaus lygties. Kondensatoriaus uždavinys <code>**talpinti krūvį (Q)**</code>, talpą matuojame faradais (F).

\begin{multline*}
i = C\dfrac{dv}{dt}
\end{multline*}


\begin{multline*}
 C = \dfrac{q}{v}
\end{multline*}

Kondensatorius yra pagamintas iš dviejų laidžių paviršių, esančių arti vienas kito. Tarp plokščių gali būti oro arba bet kokios kitos izoliacinės medžiagos. Talpos vertė priklauso nuo daugelio veiksnių, plokščių ploto, atstumo tarp plokščių (izoliatoriaus storio) ir nuo izoliacinės medžiagos fizinių savybių.

Cilindriniai kondensatoriai (poliarizuoti) yra pagaminti iš dviejų metalinių folijos plokščių, suvyniotų, kad būtų maksimaliai padidintas plokščių plotas, kad būtų pasiektos didelės talpos vertės kompaktiškoje pakuotėje.
Apskritimo formos kondensatoriai (nepoliarizuoti) yra tiesiog du metaliniai diskai, atsukti vienas į kitą, atskirti izoliatoriumi.
Reguliuojami kondensatoriai naudoja orą kaip izoliatorių. Vienas plokščių rinkinys sukasi taip, kad uždengtų daugiau ar mažiau ploto su stacionariu plokščių rinkiniu. Pavyzdžiui, radijo imtuvams derinti naudojami kintami oro kondensatoriai.
Labiausiai tikėtinas nukrypimas nuo idealaus kondensatoriaus lygties įvyksta, jei kondensatoriaus įtampa tampa tokia didelė, kad sugenda izoliacija tarp plokščių. Kai tai atsitiks, kibirkštis gali išdegti per izoliaciją ir peršokti tarp plokščių, kondensatorius <code>**sprogsta**</code>. Tikri kondensatoriai turi nominalią įtampą, kurios negalima viršyti.
Kadangi kondensatorius turi prijungimo kontaktus, jis neišvengiamai turi mažą parazitinę varžą ir induktyvumą. Parazitinis induktyvumas gali būti svarbus, jei tikimasi, kad kondensatorius sukels staigius srovės pliūpsnius, pavyzdžiui, kai jis prijungtas prie skaitmeninės lusto maitinimo. Staigus srovės padidėjimas skaitmeninei lustai reiškia, kad kondensatoriaus jungčių induktyvumas turėtų būti labai mažas.
Medžiaga, skirianti kondensatoriaus plokštes, turėtų būti izoliacinė (nepraleisti srovės). Tačiau ne visi izoliatoriai yra tobuli, todėl gali prasiskverbti mažas kiekis srovės. Atrodo, kad šios vadinamosios nuotėkio srovės teka tiesiai per kondensatorių, net jei įtampa nesikeičia.

Prijungus kondesatoriaus plokšteles prie baterijos, tos plokštelės gali talpinti krūvį (Q), tuo ir naudingi kondesatoriai. Prijungus prie baterijos neigiamas krūvis bus traukiamas link pozityvaus baterijos terminalo, o kitoje terminalo pusėje (-), neigiamas krūvis bus stumiamas toliau nuo terminalo. Kuomet neigiamas krūvis traukiasi nuo metalinės plokštelės tai padaro tą plokštelę įkrautą pozitiviu krūviu. Nes šis plokštelės dalis turi mažiau neigiamo krūvio nei teigiamo. Ploštelė kuri yra prijungta prie neigiamo terminalo tampa neigiama įkrautą nes turi daugiau neigiamo krūvio nei teigiamo. Svarbu žinoti kad abi metalo plokštelės akumuliuos vienodą kiekį krūvio, kitais žodžiais jeigu krūvis ant vienos iš plokštelių yra +2C tai ant kitos bus -2C. Dėl kad dėl kiekvienio neigiamo krūvio kuris buvo pašalintas iš vienos plokštelės toks pats kiekis krūvio perkeltas į kitą pusę.

Net jeigu ploštelių dydis ir formą būtų skirtinga jie vistiek talpintu vienoda priešinga krūvį. Šis neigiamo krūvio procesas keisti puses nevyksta visą laiką, neigiamas krūvis kuris yra traukiamas baterijos teigiamo terminalo po kurio laiko bus traukiamas prie teigiamo krūvio vienoje iš plokštelių. Galiausiai neigamas krūvis bus traukiamas vienodai į abi kriptys tuomet procesas sustoja ir akumuluotas krūvis laikosi ant plokštelių net pašalinus bateriją iš grandinės krūvis bus stabilus nes nėra kelio neigiamam krūviui pereiti į pozityviai įkrautą plokštelę. Tai iš esmės paaiškina kodėl kondesatoriaus plošteles yra atskirtos viena nuo kitos popieriaus juostelėmis, nes antraip krūvis nebūtų atskirtas ir neigiamas krūvis tiesiog būtų pastoviam rotacijos cikle uždaroje gradinėje.

Ne visi kondensatoriai gali talpinti vienoda kiekį krūvio nesvarbu prie kokios baterijos ar kito energijos šaltinio bus prijungtas. Kondensatoriaus talpa <code>**(C)**</code> kuri nusako šito įrenginio fizines galimybes talpinti krūvį nusako šį lygtis:
<br>
\begin{multline*}
C = \dfrac{Q}{V}
\end{multline*}



Talpa yra lygi taplinamam krūvio proporcijai su įtampa kondesatoriuje. Techniškai suminis krūvis kondesatoriuje yra 0, dėl to kad jis talpina vienodai teigiamo ir neigiamo krūvio. Q šioje lygtyje asociacijuoasi su krūvio dydžiu vienoje kondesatoriaus plokštelėje. Įtampa šioje lygytyje asocijuojasi su tuo kad kuomet kondesatorius talpina krūvį sukuriama įtampa (ekektrinio potencialo skirtumas) tarp dvejų kondesatoriaus plokštelių. Elektrinis potencialas yra aukštas šalia + krūvio ir yra žemas greta - krūvio. Kaip jau nagrinėjome įtampa jeigu turime teigiamo ir neigiamo krūvio šalia vienas kito bus elektrinio potencialo skirtumas tame regione, tai ką mes vadiname įtampa.  Pravartu žinoti kad jeigu baterija įkrauna kondesatorių tuomet įtampą tarp kondesatoriaus plokštelių bus tokia pati kaip ir baterijos. Kondesatoriaus talpa (C) matuojame (F) faradais.<br>
Taigi jeigu 9V baterija pilnai pakrauna 3F kondesatorių<br>

\begin{multline*}
C = \dfrac{Q}{V},\ \ 3F = \dfrac{Q}{9V}, \ \ Q = 9V x 3F = 27C
\end{multline*}

Arba jeigu 2F talpos talpina 6C krūvį <br>

\begin{multline*}
C = \dfrac{Q}{V},\ \ 2F = \dfrac{6C}{V}, \ \ V = \dfrac{6C}{2F} = 3V
\end{multline*}
<br>
Galima pamanyti jog kuo daugiau krūvio yra talpinama kondesatoriuje talpa (C) turi didėti, bet talpos vertė lieka ta pati nes kol didėja krūvis tai pat didėja ir įtampa kondensatoriuje, todėl proporcija lieka ta pati. Vienintels būdas padidinti kondensatoriaus talpa yra fiziškai pakeisti pačių plokštelių paviršiaus plotą, tarkime jeigu jos būtų didesnės arba atitrauktos toliau viena nuo kitos. Vien tik keičiant krūvį arba įtampa talpa nepasikeis.

### Kondensatoriaus talpa (papildomas medžiaga)

Įrodymas dėl <code>**𝐶=𝑄/𝑉**</code>

Isivaizduokime begalinio dydžio vienodai įkrautą plokštumą ir šios plokštumos generuojamas elektrinis laukas (elektrinio lauko jonai) yra pastovus. Šio lauko stripris, dydis 𝐸⃗ yra:<br>

\begin{multline*}
\vec{E} = 2K \pi \sigma
\end{multline*}

<code>**𝐾 - Kulono konstanta**</code><br>
<code>**𝜎 - plokštumos krūvio tankis**</code>


\begin{multline*}
\sigma = \dfrac{Q}{A}
\end{multline*}

Jeigu plokštumos plotis yra begalinis todėl ir krūvis bus begalinis. Bet jeigu plotas yra ribotas tuomet arti plokštumos centro mes turime krūvio vidurkį kuris representuoja vienodai įkrautą plokštumą.

{iliustracija}<br>
Jeigu turime dvi plokštumas kurios <code>**turi plotą A ir atstumą tarp jų d**</code>. Tarp šių plokštelių elektrinis laukas bus kombinuotai sugeneruotas abiejų plokštumų jie abu bus pastovus ties centru. d yra mažas. ... . Abu laukai bus vienkrypčiai ir vienodo dydžio, išsipūsdami ant plokštumos kraštų. Uždavinys išsiaiškinti ryšį tarp įtampos tarp šių plokštumų ir ploto bei atstumo tarp jų. Mes žinome kad elektrinis laukas sugeneruotos kurios nors iš įkrautų plokštumų yra:<br>

\begin{multline*}
\vec{E} = 2K \pi \dfrac{Q}{A} + 2K \pi \dfrac{Q}{A} = 4K \pi \dfrac{Q}{A}
\end{multline*}

<code>**PASTABA:sudėtis tarp laukų nes laukai yra vienkrypčiai.**</code><br>
Siekiant išsiaiškint įtampos skirtuma tarp dvejų taškų {illiustracija}. Įtampos skirtumas yra elektrinio potencialo energijos per krūvį (jeigu krūvis yra lokalizuotas, tai reiškia jeigu kruvis yra vienoje versus kitoje pozicijoje). Tarkime (bet taip nėra) pozityvus jonas nori patekti į neigiamai įkrautą plokštelės pusę (elektros tekmė - neigiamo krūvio judėjimas). Tai įtampa gali sakyti yra matavimas nusakantis kaip stipriai krūvis nori patekti kitur, jeigu tik turi kaip (jeigu tarp plokštelių yra oras arba vakumas krūvis negali judėti), tai nusako tarytumei elektrinį spaudimą.
