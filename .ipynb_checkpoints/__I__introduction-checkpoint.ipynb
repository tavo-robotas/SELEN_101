### Krūvis

Elektros sąvoka kylo iš gamtos stebėjimo. Mes pastebime jėgą tarp objektų, kuri, kaip ir gravitacija, veikia per atstumą. Šios jėgos šaltinis buvo pavadinta kruviu nuo to ir prasideda istorija apie srovę. 

Pastebimas elektrinės jėgos dalykas yra tai, kad ji yra didelė, daug didesnė už gravitacijos jėgą. Tačiau skirtingai nuo gravitacijos, yra dviejų tipų elektros krūviai teigiamas <code>**(+)**</code> ir neigiamas <code>**(-)**</code>. Priešingi krūvių tipai pritraukia ir panašūs kruvių tipai stumia. Gravitacija turi tik vieną tipą: ji tik traukia, niekada nestumia.

Krūvio nešėjas yra įkrauta (neigiamą arba teigiamą potencialą turinti) dalelė kvantiniame lauke. Neigiamo krūvio dalelė [elektronas](https://en.wikipedia.org/wiki/Electron), tik kur mums jų gauti ?

Mus domina išorinė vieno elektrono orbita šiame vario (Cu) atome. Išorinė orbitalė turi tik vieną taip vadinama valentinį elektroną kas daro šį eletroną lengvai sužadinama išlėkti iš orbitos, nes jis turi santykinai silpnus ryšius su savo branduoliu. Todėl galime (induce conduction) - sukelti procesą, kurio metu šiluma arba elektra tiesiogiai perduodama per medžiagą, kai skiriasi temperatūrų arba elektrinio potencialo skirtumas tarp gretimų sričių, medžiagai nejudant. 
Kai metalo atomų krūva yra kartu, jie dalijasi savo išoriniais elektronais vienas su kitu, sukurdami elektronų "spiečius", nesusijusius su konkrečiu branduoliu. Labai maža elektrinė jėga gali priversti elektronų spiečius judėti. Todėl varis, auksas, sidabras ir aliuminis dėl šios valentinių elektronų savybės yra geri laidininkai, taip pat ir sūrus vanduo dėl ištirpusio NaCl.

Srovė matuojama amperais (A). Vienas amperas tai srovė apibrėžiama kaip vienas <code>**[kulonas](https://lt.wikipedia.org/wiki/Kulonas)**</code> per sekundę. Tai reiškia, kad vieną amperą galima įsivaizduoti kaip maždaug 6.242E18 elektronų, praeinančių per laidą konkrečioje vietoje per vieną sekundę.
