## Nuoseklus rezistoriai

Komponentai yra sujungti nuosekliai, jei jie yra sujungti galais taip kaip žemiau esančiame paveikslėlyje. Atliksime kelis darbus su rezistoriais, kad nagrinėti nuoseklaus ryšio savybes. Svarbu įsiminti kad rezistoriai yra sujungti nuosekliai, kai jie yra sujungti vienas su kitu ir tarp nėra kitų laidų, išsišakojančių iš mazgų tarp komponentų.

Nuoseklus rezistoriai turi tą pačią srovę. Toliau pateiktame paveikslėlyje esantys rezistoriai nėra nuoseklūs. Tarp rezistorių yra papildomų šakų, vedančių nuo mazgų. Jei šios šakos teka srovę (oranžinės rodyklės), tada 𝑅1, 𝑅2, 𝑅3 nesidalykite ta pačia srove.

Taip atrodo grandinė su nuosekliais rezistoriais:

$$I$$

Įtampos šaltinis 𝑉𝑆 prijungtas prie nuoseklios rezistoriaus grandinės. Įtampa 𝑉𝑆 yra kažkokia pastovi reikšmė, bet mes dar nežinome kokia yra srovė 𝐼 or kaip 𝑉𝑆 dalijasi tarp trijų rezistorių. Yra du dalykai kuriuos mes žinome 
 - Visų trejų rezisotorių įtampa turi sumuotis į 𝑉𝑆;
 - Srovė 𝐼 eina per visus rezistorius,
 
Turėdami šiek tiek žinių ir Omo dėsnį, galime parašyti šias išraiškas: 

\begin{multline*}
V_S = V_{R1} + V_{R2} + V_{R3}
\end{multline*}

\begin{multline*}
V_{R1} = i \cdot R_1
\end{multline*}

\begin{multline*}
V_{R2} = i \cdot R_2
\end{multline*}

\begin{multline*}
V_{R3} = i \cdot R_3
\end{multline*}

To pakanka norint sujungti lygtis į:

\begin{multline*}
V_{S} = i \cdot R_{1} + i \cdot R_{2} + i \cdot R_{3}
\end{multline*}

Įvertinus srovę ir surinkus rezistorius vienoje vietoje gauname:

\begin{multline*}
V_{S} = i (R_1 + R_2 + R_3)
\end{multline*}

Kadangi žinome 𝑉𝑆, randame nežinomajį 𝑖

\begin{multline*}
i = \dfrac{V_{S}}{(R_1 + R_2 + R_3)}
\end{multline*}

Tai atrodo kaip Omo dėsnis vienam rezistoriui, išskyrus tai, kad nuoseklus rezistoriai rodomi kaip suma. Darome išvadą kad nuoseklių rezistorių bendra varža yra atskirų rezistorių varžų suma. Galime įsivaizduoti naują didesnį rezistorių, atitinkantį nuoseklių rezistorių sumą. Ji yra lygiavertė ta prasme, kad tam tikram 𝑉𝑆 teka ta pati srovė 𝑖.

\begin{multline*}
R_{\text{nuoseklus}} = R_1 + R_2 + R_3
\end{multline*}

Galite išgirsti žargono posakį: Iš įtampos šaltinio požiūrio taško tie trys nuoseklus rezistoriai atrodo kaip vienas didesnis lygiavertis jiems rezistorius. Tai reiškia, kad srovė i, kurią teikia įtampos šaltinis, abiem atvejais yra tokia pati.

Jei turite keletą nuosekliai sujungtų rezistorių, bendra lygiavertė varžos forma yra:

\begin{multline*}
R_{\text{nuoseklus}} = R_1 + R_2 + \ ... \ + R_n
\end{multline*}


### Įtampa pasiskirstymas tarp nuosekliai sujungtų rezistorių

Mes jau nustatėme srovę i per nuoseklias jungtys, todėl beliko išsiaiškinti atskirų rezistorių įtampą. Padarome tai atskiriems rezistoriams taikant Omo dėsnį.

\begin{multline*}
V_{R1} = i \cdot R_1
\end{multline*}

\begin{multline*}
V_{R2} = i \cdot R_2
\end{multline*}

\begin{multline*}
V_{R1} = i \cdot R_3
\end{multline*}

Tai tampa aiškai, jei naudosime pavyzdį su realiais skaičiais. Raginu jus tai išbandyti patiems:
 - Išspręskite kokia yra šios gradinės srovė 𝑖 ir kokios yra trijų rezistorių įtampos.
 - Parodykite atskirų rezistorių įtampų sumą kuri susideda į bendrą grandinės šaltinio įtampą.

Sprendinio žingnsiai:
 - Susirasti R(nuoseklus) reikšmę;
 - Pagal Omo dėsnį susirasti srovę 𝑖;
 - Vėl naudojant Omo dėsnį susirasti individualias įtampas per varžas;
 - Patikrinti ar įtampos susimuoja į tai kokia turi būti atsižvlegiant į šaltinio įtampą.

- R = 2KΩ
- V = IR, I = V/R, 10V/2000Ω = 0.005A = 5mA
- R1: 0.005A * 300Ω  = 1.5V
- R2: 0.005A * 500Ω  = 2.5V
- R3: 0.005A * 1200Ω = 6V

<code>**Prisimink:**</code>
- Nuoseklus rezistoriai turi tą pačią srovę.
- Nuoseklus  rezistoriai susideda,
\begin{multline*}
R_{\text{nuoseklus}} = R_1 + R_2 + \ ... \ + R_n
\end{multline*}
- Įtampa paskirstyto tarp nuoseklių rezistorių, o didžiausia įtampa buna didžiausios varžos rezistoriuje.

## Lygiagretus rezistoriai

Komponentai yra lygiagretus, jei jie du mazgai sujungti kaip šiame pavyzdyje kuriame rezistoriai yra sujungti lygiagrečiai, kai jų du gnybtai jungiasi prie tų pačių mazgų:

Šiame skyriuje nagrinėsime lygiagrečiai sujungtus rezistorius, kad atskleistume lygiagrečios jungties savybes. Vėlesnese paskaitose aptarsime nuoseklius ir lygiagrečius kondensatorius ir induktorius. Žemiau esančoame paveikslėlyje R1, R2, R3 yra lygiagrečiai sjungti rezistoriai. Du paskirstyti mazgai pavaizduoti dviem horizontaliomis linijomis.

Lygiagrečiai rezistoriai turi tą pačią įtampą savo gnybtuose. Toliau pateiktame paveikslėlyje esantys rezistoriai nėra lygiagrečiai. Yra papildomų komponentų (oranžinės dėžutės), suskaidančių bendrus mazgus tarp rezistorių. Ši grandinė turi keturis atskirus mazgus, todėl R1, R2, R3 neturi tos pačios įtampos

Išsiaiškinti sroves ir įtampas su lygiagrečius rezistoriais yra šiek tiek sudėtingiau nei su nuosekliais. Čia yra grandinė su lygiagrečiais rezistoriais.

Srovės šaltinis 𝐼𝑆 nukreipia srovę 𝑖 link R1, R2, R3. Žinome, kad srovės 𝑖 reikšmė yra tam tikra konstanta, pastovi. Bet dar nežinome įtampos 𝑣 arba kaip 𝑖 padalijama į tris rezistorių sroves. Yra du dalykai kuriuos žinome:
 - Trijų rezistorių srovės turi susisumuoti į 𝑖;
 - Įtampa 𝑣 pasirodo visuose trijuose rezistoriuose.
 
Turėdami tik šiek tiek žinių ir Omo dėsnį, galime parašyti šias išraiškas taip:

\begin{multline*}
i = i_{R1} + i_{R2} + i_{R3}
\end{multline*}

\begin{multline*}
v = i_{R1} \cdot R1 
\end{multline*}

\begin{multline*}
v = i_{R2} \cdot R2 
\end{multline*}

\begin{multline*}
v = i_{R3} \cdot R3
\end{multline*}

To pakanka norint pertvarkykite tris Omo dėsnio išraiškas, kad išspręstumėte srovę pagal įtampą ir varžą:

\begin{multline*}
i_{R1} = \dfrac{v}{R1}
\end{multline*}


\begin{multline*}
i_{R2} = \dfrac{v}{R2}
\end{multline*}


\begin{multline*}
i_{R3} = \dfrac{v}{R3}
\end{multline*}

Pakeiskite šias išraiškas į srovių sumą ir išskirkite bendrą terminą:

\begin{multline*}
i = \dfrac{v}{R1} + \dfrac{v}{R2} +\dfrac{v}{R3}
\end{multline*}

\begin{multline*}
i = v(\dfrac{1}{R1} + \dfrac{1}{R2} +\dfrac{1}{R3})
\end{multline*}

Mes žinome 𝑖 (tai yra dabartinio šaltinio savybė), todėl galime išspręsti 𝑣:

\begin{multline*}
v = i \dfrac{1}{(\dfrac{1}{R1} + \dfrac{1}{R2} +\dfrac{1}{R3})}
\end{multline*}

Išvada kad lygiagrečiai veikiančių rezistorių bendra varža yra atskirų rezistorių atvirkštinių verčių sumos atvirkštinė vertė. Tai skamba sudėtingai, bet prieš tai padarysime ką nors paprastesnio.

### Lygiavertis lygiagretus rezistorius

Ankstesnė lygtys rodo, kad galime apibrėžti naują rezistorių atitinkantį lygiagrečius rezistorius. Naujasis rezistorius yra lygiavertis ta prasme, kad esant tam tikrai i, turės tą pačią įtampa v.

\begin{multline*}
R_{\text{lygiagretus}} = \dfrac{1}{(\dfrac{1}{R1} + \dfrac{1}{R2} +\dfrac{1}{R3})}
\end{multline*}

Lygiagretusis rezistorius yra atvirkštinių dydžių sumos atvirkštinė vertė. Šią lygtį galime parašyti kitaip, pertvarkydami taip:

\begin{multline*}
\dfrac{1}{R_{\text{lygiagretus}}} = \dfrac{1}{R1} + \dfrac{1}{R2} +\dfrac{1}{R3}
\end{multline*}

\begin{multline*}
v = i R_{\text{lygiagretus}}
\end{multline*}

Srovės šaltinio požiūros taško atžvilgiu lygiagretus rezistorius R nesiskiria nuo trijų lygiagrečių rezistorių, nes abiejose grandinėse v yra tas pats.

Jei lygiagrečiai turite kelis rezistorius, bendra lygiagrečios varžos forma yra:

\begin{multline*}
\dfrac{1}{R_{\text{lygiagretus}}} = \dfrac{1}{R1} + \dfrac{1}{R2} + \ ...\ +\dfrac{1}{Rn}
\end{multline*}

### Srovė paskirstoma tarp rezistorių lygiagrečiai

Mes nustatėme įtampą v lygiagrečiame jungtyje, todėl belieka išsiaiškinti sroves per atskirus rezistorius. Padarykime tai atskiriems rezistoriams taikydami Omo dėsnį.

\begin{multline*}
V = i_{R1} \cdot R1
\end{multline*}

\begin{multline*}
V = i_{R2} \cdot R2
\end{multline*}

\begin{multline*}
V = i_{R2} \cdot R2
\end{multline*}

Tai tampa daug aiškiau jei spręsite pavyzdį su realiais skaičiais. Raskite įtampą v ir sroves per tris rezistorius.
Parodykite, kad atskiros rezistorių srovės sudaro i.

Sprendimo žingsniai yra šie:
1. Raskite lygiagrečiąją varžą R lygiagretus
2. Tada raskite įtampą v, naudodami Omo dėsnį.
3. Tada suraskite atskiras sroves, dar kartą naudodami Omo dėsnį.
4. Patikrinkite, ar rezistorių srovės susideda į bendrą suma ir yra tokia kokia turėtų būti.

- R_lygiaretus =  1/ (0.02 + 0.01 + 0.002) = 1/0.032 = 31.25Ω
- V = IR, 0.1A * 31.25Ω =  3.125V
- iR1 = V/R1 = 3.125V/50Ω = 0.0625A = 62.5mA
- iR2 = V/R2 = 3.125V/100Ω = 0.03125A = 31.25mA
- iR3 = V/R3 = 3.125V/500Ω = 0.00625A = 6.25mA 

### Specialus atvejis – du rezistoriai lygiagrečiai

Du lygiagrečiai sujungti rezistoriai turi tokią varžą:

\begin{multline*}
R_{\text{parallel}} = \dfrac{1}{{(\dfrac{1}{R1} + \dfrac{1}{R2}})}
\end{multline*}

Galima šiek tiek keisti lygtį, kad pašalintumėte atvirkštinius dydžius ir sugalvotumėte kitą išraišką tik su viena trupmena.  Tai algebrinių operacijų seka kuri duoda šią išraiška:

\begin{multline*}
R_{\text{parallel}} = \dfrac{R1 \cdot R2}{R1 + R2}
\end{multline*}

### Specialus atvejis – lygiagrečiai du vienodi rezistoriai

Jei du lygiagrečiai sujungti rezistoriai turi tą pačią vertę:

\begin{multline*}
R_{\text{parallel}} = \dfrac{1}{2}R
\end{multline*}

Du lygiagrečiai sujungti identiški rezistoriai jų lygiavertę varžą yra pusę bet kurio rezistoriaus vertės. Srovė pasiskirsto vienodai tarp dviejų.

## Lygiagretus laidumas

Mes išvedėme lygtį norėdami sujungti lygiagrečius rezistorius į vieną lygiavertį rezistorių:

\begin{multline*}
R_{\text{lygiagretus}} = \dfrac{1}{(\dfrac{1}{R1} + \dfrac{1}{R2} +\dfrac{1}{R3})}
\end{multline*}

Tai gana sudėtinga išraiška, su 1/R terminais, įterptais į kitą atvirkštinius dydžio išraišką. Yra alternatyvus būdas išspręsti šią problemą, naudojant laidumo sąvoką. Omo dėsnis V = IR, apibrėžia varžą kaip įtampos ir srovės santykį.

\begin{multline*}
R = \dfrac{v}{i}
\end{multline*}

Laidumas yra atvirkštinis šios išraiškos variantas. Tai srovės ir įtampos santykis:

\begin{multline*}
G = \dfrac{i}{v}
\end{multline*}

Todėl yra dar vieną būdas parašyti Omo dėsnį:

\begin{multline*}
i = vG
\end{multline*}

Laidumo vienetas yra siemens, sutrumpintai S. Jis pavadintas Vokietijos pramoninės elektronikos ir telekomunikacijų įmonės įkūrėjo Wernerio von Siemenso vardu. Naudojant laidumą, o ne varžą tam pačiam fiziniam objektui, tiesiog pabrėžiamas kitas jo elgesio aspektas. Atsparumas sumažina arba trukdo srovės srautui, o laidumas leidžia srovei praeiti. Terminai yra du tos pačios idėjos aspektai. 100 Ω rezistorius yra toks pat kaip laidumas: 0.01S = 1/100 Ω

### Lygiagretus laidumas

Šioje dalyje šiek tiek pakartosime lygiagrečių rezistorių analizę, tačiau šį kartą, užuot pavadinę kiekvieną komponentą rezistoriumi, pavadinsime jį laidumu. Lygiagrečio laidumo rezultatas bus labai panašus į nuoseklius rezistorius. Čia yra grandinė su lygiagrečiu laidumu. Mes analizuosime šią grandinę naudodami laidumo savoką ir Omo dėsnio laidumo formą i = vG.

Srovės i reikšmė yra tam tikra konstanta. Mes dar nežinome v arba kaip i per laidumą skyla į tris sroves. Mes žinome du dalykus:
- Trys laidumo srovės susideda i
- Įtampa v atsiranda visuose trijuose laidumuose.

Su šiek tiek žinių ir Omo dėsnio laidumo formą, galime parašyti šias išraiškas taip:

\begin{multline*}
i = i_{G1} + i_{G2} + i_{G3}
\end{multline*}

\begin{multline*}
i_{G1} = v \cdot G1
\end{multline*}

\begin{multline*}
i_{G2} = v \cdot G2
\end{multline*}

\begin{multline*}
i_{G3} = v \cdot G3
\end{multline*}

To pakanka norint padaryti lygčių sujungimą:

\begin{multline*}
i = v \cdot G1 + v \cdot G2 + v \cdot G3
\end{multline*}


\begin{multline*}
i = v (G1 + G2 + G3)
\end{multline*}

Tai atrodo taip pat kaip Omo dėsnis vienam laidumui, kai lygiagretus laidumai rodomi kaip suma. Lygiagrečių laidumų atveju bendras laidumas yra atskirų laidumų suma. Atkreipkite dėmesį, kaip tai atrodo kaip nuosekliųjų rezistorių formulė. Laidumas lygiagrečiai yra kaip varžos nuosekliai.
